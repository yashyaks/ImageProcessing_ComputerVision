```python
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

image_path = 'images/test.jpg'
img = Image.open(image_path)

img_array = np.array(img)

plt.figure(figsize=(8, 8))
plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

kernel_hp = np.array([[-1, -1, -1],
                     [-1,  8, -1],
                     [-1, -1, -1]])
img_hp = cv2.filter2D(img_array, -1, kernel_hp)

plt.subplot(1, 3, 2)
plt.imshow(img_hp)
plt.title('High Pass Filter')
plt.axis('off')

A = 1.1
kernel_hb = np.array([[-1, -1, -1],
                      [-1,  9*A-1, -1],
                      [-1, -1, -1]])
img_hb = cv2.filter2D(img_array, -1, kernel_hb)

plt.subplot(1, 3, 3)
plt.imshow(img_hb)
plt.title(f'High Boost Filter (A={A})')
plt.axis('off')

plt.show()
```
